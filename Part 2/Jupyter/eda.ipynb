# Cleaning the data
Only keep entries where GUID exists in both tables

```python
common_guids = avf_tests.merge(host_tests, on="Guid", how="inner")["Guid"].unique()
common_guids.size
```

```python
# Define function to determine final result status
def resolve_result(results):
    return "PASS" if all(results == "PASS") else "FAIL"

# Function to process a dataframe
def process_dataframe(df):
    df_filtered = (
        df[df["Guid"].isin(common_guids)]  # Filter by given GUIDs
        .groupby("Guid", as_index=False)  # Group by Guid
        .agg({
            "Service\\Protocol Name": "first",  # Keep first occurrence
            "Result": resolve_result,  # Apply custom logic for Result
            "Title": "first"  # Keep first occurrence
        })
    )
    
    # Split "Service\Protocol name" into two columns: Group and Protocol
    df_filtered[["Service", "Protocol"]] = df_filtered["Service\\Protocol Name"].str.split("\\", expand=True)
    return df_filtered

# Process both dataframes
avf_tests_filtered = process_dataframe(avf_tests)
host_tests_filtered = process_dataframe(host_tests)

# Display results
avf_tests_filtered
```

# Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 1: Filter only FAILED tests
avf_failures = avf_tests_filtered[avf_tests_filtered["Result"] == "FAIL"].groupby("Service").size().reset_index(name="Failures (VM)")
host_failures = host_tests_filtered[host_tests_filtered["Result"] == "FAIL"].groupby("Service").size().reset_index(name="Failures (Host)")

# Step 2: Merge results on "Group" (Service)
failures_df = pd.merge(avf_failures, host_failures, on="Service", how="outer").fillna(0)

# Step 3: Convert to long format for Seaborn
failures_df = failures_df.melt(id_vars="Service", var_name="Test Type", value_name="Failure Count")

# Step 4: Plot the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x="Service", y="Failure Count", hue="Test Type", data=failures_df)

# Formatting
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for readability
plt.xlabel("Service")
plt.ylabel("Number of Failed Tests")
plt.title("Comparison of Failed Tests per Service")
plt.legend(title="Test Source")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.yticks(np.arange(0, 20, step=2))

# Show and save plot
plt.tight_layout()
plt.savefig("comparison.png", facecolor='w')
plt.show()
```

```python
# Merge test results on 'Guid' (unique test identifier)
merged_tests = avf_tests_filtered.merge(host_tests_filtered, on="Guid", suffixes=("_avf", "_host"))

# Filter where AVF test failed but Host test passed
fail_in_avf = merged_tests[(merged_tests["Result_avf"] == "FAIL") & (merged_tests["Result_host"] == "PASS")]

# Count occurrences per 'Service/Protocol Name'
fail_counts = fail_in_avf["Service\\Protocol Name_avf"].value_counts()

# Plot bar chart
plt.figure(figsize=(12, 6))
fail_counts.plot(kind="bar", color="red")
plt.xlabel("Service/Protocol Name")
plt.ylabel("Number of Tests FAIL in AVF, but PASS in Host)")
plt.title("Tests Failed in AVF but Passed in Host")
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show and save plot
plt.tight_layout()
plt.savefig("failedinvm.png", facecolor='w')
plt.show()
fail_in_avf
```
